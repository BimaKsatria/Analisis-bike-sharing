```python
hours_aggregated = hours_df.groupby("dteday").agg({
    "casual": "sum",
    "registered": "sum",
    "count": "sum"
}).reset_index()

all_df = pd.merge(
    left=days_df,
    right=hours_aggregated,
    how="left",
    on="dteday"
)

all_df.head()
```

```python
# Ganti label 'season' dengan nama musim yang lebih deskriptif
all_df['season'] = all_df['season'].replace({
    1: 'Musim Semi',
    2: 'Musim Panas',
    3: 'Musim Gugur',
    4: 'Musim Dingin'
})

season_data = all_df.season
print(season_data)
```

**Insight:**
- Menggabungkan data days.csv dan hours.csv menjadi all_data.csv
- Mengganti label sesason menjadi deskriptif agar mudah di pahami

## Visualization & Explanatory Analysis

### Pertanyaan 1: - Apakah ada peningkatan jumlah pengguna setiap bulan?

```python
# Tambahkan kolom 'month' untuk memisahkan data berdasarkan bulan
all_df['dteday'] = pd.DatetimeIndex(all_df['dteday']).strftime('%Y-%m')

# Agregasi data untuk menghitung jumlah pengguna per bulan
monthly_users_df = all_df.groupby('dteday').agg({
    "casual_y": "sum",     # Jumlah total pengguna casual per bulan
    "registered_y": "sum"  # Jumlah total pengguna terdaftar per bulan
}).reset_index()

# Hitung total pengguna (casual + registered) per bulan
monthly_users_df['total_users'] = monthly_users_df['casual_y'] + monthly_users_df['registered_y']

# Menghitung perubahan pengguna dari bulan ke bulan
monthly_users_df['user_change'] = monthly_users_df['total_users'].diff()

# Hitung persentase perubahan
monthly_users_df['user_change_pct'] = (monthly_users_df['user_change'] / monthly_users_df['total_users'].shift(1)) * 100

# Optional: Plot untuk visualisasi
fig, ax = plt.subplots(figsize=(10,6))
ax.plot(monthly_users_df['dteday'], monthly_users_df['total_users'], marker='o')
ax.set(title='Jumlah Pengguna Sepeda per Bulan', xlabel='Bulan', ylabel='Jumlah Pengguna')
plt.xticks(rotation=90)
plt.show()
```

### Pertanyaan 2: Pada musim apa pengguna sepeda mengalami kenaikan?

```python
# Mengelompokkan data berdasarkan musim (season) dan kondisi cuaca (weathersit)
users_by_season_df = all_df.groupby('season').agg({
    "casual_y": "sum",     # Jumlah total pengguna casual per musim
    "registered_y": "sum"  # Jumlah total pengguna terdaftar per musim
}).reset_index()

# Hitung total pengguna (casual + registered) per musim
users_by_season_df['total_users'] = users_by_season_df['casual_y'] + users_by_season_df['registered_y']


# Visualisasi untuk membandingkan jumlah pengguna per musim
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(10,6))
ax.bar(users_by_season_df['season'], users_by_season_df['total_users'], color=['#D3D3D3', '#90CAF9', '#D3D3D3', '#D3D3D3'])
ax.set(title='Jumlah Pengguna Sepeda per Musim', xlabel='Musim', ylabel='Jumlah Pengguna')
plt.show()
```
